```python


plt.figure(figsize=(20,20))
raw_image = cv2.imread('./sample_img/C7.jpg')

plt.subplot(221)
image = cv2.cvtColor(raw_image,cv2.COLOR_BGR2RGB)
plt.imshow(image)

plt.subplot(222)
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,'gray')

plt.subplot(223)
cropped_image = gray[132:132+284, 296:296+284]
plt.imshow(cropped_image,'gray')

plt.subplot(224)
gray_sm = cv2.bilateralFilter(cropped_image,7,20,20)
outline = cv2.Canny(gray_sm,50,70)
plt.imshow(outline,'gray')
```
